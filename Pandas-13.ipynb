# Membuat kolom baru dengan for loop dan f-string

```python
d = {'penjual':['bejo','tejo','wati','bejo','cecep','tejo','wati','bejo'],
     'barang':['monitor','monitor','keyboard', 'mouse','keyboard','monitor','laptop','monitor']}
df = pd.DataFrame(d)
df
```

```python
cols = ['penjual', 'barang'] #berasosiasi dengan kolom pada Data Frame

for cols in cols:
  df[f'count_tiap_{cols}'] = df.groupby(cols).cumcount()+ 1 #untuk melakukan pengelompokkan berdasarkan nilai jualnya dari setiap kolom
df
```

# Seleksi Baris dengan between  

```python
n_rows = 10
n_cols = 5

cols = tuple('ABCDE')

df=pd.DataFrame(np.random.randint(1,10, size=(n_rows,n_cols)), columns = cols)
df
```

```python
df[df['A'].between(2,5)] #agar nilai kolom A berada dalam jangkauan 2 dan 5, between bersifat insklusif, maka batasan nilai disertakan
```

```python
df[df['A'].between(2,5, inclusive=False)] #untuk menjadikan between bersifat ekslusif, sehingga batasan nilai tidak ikut disertakan
```

# Transformasi kolom menjadi baris

```python
d ={
'kode_area':[123,456,789,321],
'pabrik':[4,2,5,0],
'gudang':[2,4,7,3],
'toko':[64,32,15,24]
}

df = pd.DataFrame(d)
df
```

```python
df = df.melt(id_vars='kode_area',
             var_name='jenis_bangunan',
             value_name='jumlah') #digunakan untuk merubah kolom menjadi baris

df
```
